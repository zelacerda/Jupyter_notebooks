```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

### Using make_blobs to create data clusters

```python
from sklearn.datasets.samples_generator import make_blobs

x, y = make_blobs(n_samples=200,
                  n_features=2,
                  centers=2,
                  cluster_std=1.0,
                  center_box=(-10.0, 10.0),
                  shuffle=True,
                  random_state=None)

plt.scatter(x[:,0], x[:,1], c=y, cmap=plt.cm.Blues)
```

### Load datasets in sklearn

```python
from sklearn.datasets import load_iris

iris = load_iris()

x_index = 2
y_index = 3

plt.figure(figsize=(8,8))
plt.scatter(iris.data[:,x_index], iris.data[:,y_index], c=iris.target, cmap=plt.cm.RdYlBu)
plt.xlabel(iris.feature_names[x_index])
plt.ylabel(iris.feature_names[y_index])
```

### Using linear regression

```python
from sklearn.linear_model import LinearRegression

# Coeficients of linear equation (y = a*x + b)
a = 3
b = 10

# Number of samples
n = 100

# Generating random data
x = 30 * np.random.random(n)
y = a * x + b + 5 * np.random.normal(size=x.shape) #Introducing some noise

# Generating linear model
clf = LinearRegression()
clf.fit(x[:, None], y)

x_new = np.linspace(0, 30, 100)
y_new = clf.predict(x_new[:, None])

print clf.coef_[0], clf.intercept_

plt.figure(figsize = (10,5))
plt.scatter(x,y)
plt.plot(x_new, y_new)
```
